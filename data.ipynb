```python
# load data function with filename argument
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data

data = load_data("store_data.json")
print(data)
print(type(data))
```

## Assignment 1: Data Cleaning & Structuring

Issues identified:
1. Inconsistent data (mixed type ratings)
2. Missing values (age)
3. Duplicate user entries

```python
def clean_data(data):
    text_to_num = {"one": 1, "two":2, "three": 3, "four": 4, "five": 5} 
    cleaned_data = []
    unique_users = set()

    for user in data:
        # clean ratings - data consistency
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # handle missing value
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        # handle duplicate value (deduplication)
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)
        
    return cleaned_data
```

## Assignment 2: Getting Meaningful Insights

Questions:
1. What is the average rating?
2. What percentage of users gave a poor rating (< 3)?

```python
def get_insights(data):

    # average rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])

    print(f"avg rating = {tot_rating/len(data)}")

    # percentage of users with poor rating
    poor_rating = 0
    for user in data:
        if float(user["rating"]) < 3:
            poor_rating += 1

    print(f"Percentage of users with poor ratings = {poor_rating/len(data) *100}")
```

```python
get_insights(data)
```

## Assignment 3: Product Recommendation System (Rule-Based)

Rules:
- If rating ≥ 4 → Recommend same brand
- If rating < 4 → Recommend different brand

```python
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]

        if (float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"
            
        recommendations.append(curr_recomm)
    return recommendations
```

```python
get_recommendations(data)
```
